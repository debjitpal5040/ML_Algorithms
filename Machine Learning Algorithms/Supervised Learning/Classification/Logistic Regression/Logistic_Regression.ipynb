# Logistic Regression


## Importing the necessary libraries and packages


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from joblib import dump
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    accuracy_score,
    auc,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve,
)
from sklearn.model_selection import train_test_split
```

## Importing the csv file as a dataFrame


```python
df = pd.read_csv("../../../../Datasets/SocialNetworkAds.csv")
df.head()
```

## Exploratory Data Analysis


Formal information about the dataset

Description of the dataset


```python
df.describe()
```

Checking if any null value exists


```python
df.isnull().sum()
```

```python
df.columns
```

Correlations between the variables


```python
df.corr()
```

## Now taking 2 parts from the dataset

### X: which contains all columns except the <b>target</b> column<br>Y: which contains only the <b>target</b> column


```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

## Split the <b>X</b> and <b>y</b> Dataframes for Test and Training


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

## Implementation of the Model


```python
lr = LogisticRegression(max_iter=10000)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

## Performance of the Model


```python
lr_con_matrix = confusion_matrix(y_test, y_pred)
lr_acc = accuracy_score(y_test, y_pred)
print("Confusion Matrix", "\n", lr_con_matrix, "\n")
print("Accuracy of Logistic Regression :", lr_acc * 100, "\n")
print(
    "Classification Report",
    "\n",
    classification_report(y_test, y_pred, zero_division=1),
)
```

## Plotting the Confusion Matrix


```python
fig, ax = plt.subplots(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
ConfusionMatrixDisplay(lr_con_matrix).plot(ax=ax)
plt.title("Confusion Matrix")
plt.savefig("Images/Confusion_Matrix_Logistic_Regression.png", bbox_inches="tight")
```

## Plotting the ROC curve


```python
y_score = lr.predict_proba(X_test)[:, 1]
fpr, tpr, threshold = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
RocCurveDisplay(
    fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name="Logistic Regression"
).plot(ax=ax)
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.plot([0, 1], [0, 1], "b--")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.savefig("Images/ROC_Curve_Logistic_Regression.png", bbox_inches="tight")
```

## Computing the AUC of ROC


```python
print(
    "Area Under the Receiver Operating Characteristic Curve :",
    roc_auc_score(y_test, y_score),
)
```

# Saving the model


```python
dump(lr, "../../../../Models/logisticRegression.joblib")
```
