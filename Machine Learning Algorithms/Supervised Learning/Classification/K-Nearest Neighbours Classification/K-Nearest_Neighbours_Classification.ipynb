# K-Nearest Neighbours Classification


## Importing the necessary libraries and packages


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from joblib import dump
from sklearn.metrics import (
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    accuracy_score,
    auc,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve,
)
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

## Importing the csv file as a dataFrame


```python
df = pd.read_csv("../../../../Datasets/SocialNetworkAds.csv")
df.head()
```

## Now taking 2 parts from the dataset

### X: which contains all columns except the <b>target</b> column<br>Y: which contains only the <b>target</b> column


```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

## Split the <b>X</b> and <b>y</b> Dataframes for Test and Training


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

## Finding the best value of K for KNN Classification


```python
n = X_train.shape[0]
rmse_val = []
for K in range(1, int(n**0.5) + 1):
    model = KNeighborsClassifier(n_neighbors=K)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    rmse_val.append(100 * accuracy_score(y_test, pred))  # type: ignore
plt.figure(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
plt.plot(
    range(1, int(n**0.5) + 1),
    rmse_val,
    linestyle="dashed",
    color="blue",
    marker="o",
    markerfacecolor="red",
)
plt.xticks(np.arange(1, int(n**0.5) + 1, 2))
plt.yticks(np.arange(85, 100, 2))
plt.xlabel("Value of K")
plt.ylabel("Accuracy (%)")
plt.title("Value of K vs Accuracy")
plt.savefig("Images/Value_of_K_vs_Accuracy.png", bbox_inches="tight")
```

## Implementation of the Model


```python
knnc = KNeighborsClassifier(n_neighbors=8)
knnc.fit(X_train, y_train)
y_pred = knnc.predict(X_test)
```

## Performance of the Model


```python
knnc_con_matrix = confusion_matrix(y_test, y_pred)
knnc_acc = accuracy_score(y_test, y_pred)
print("Confusion Matrix", "\n", knnc_con_matrix, "\n")
print("Accuracy of K-Nearest Neighbours Classification :", knnc_acc * 100, "\n")  # type: ignore
print("Classification Report", "\n", classification_report(y_test, y_pred))
```

## Plotting the Confusion Matrix


```python
fig, ax = plt.subplots(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
ConfusionMatrixDisplay(knnc_con_matrix).plot(ax=ax)
plt.title("Confusion Matrix")
plt.savefig(
    "Images/Confusion_Matrix_K-Nearest_Neighbours_Classification.png",
    bbox_inches="tight",
)
```

## Plotting the ROC curve


```python
y_score = knnc.predict_proba(X_test)[:, 1]  # type: ignore
fpr, tpr, threshold = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
RocCurveDisplay(
    fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name="K-Nearest Neighbours Classifier"
).plot(ax=ax)
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.plot([0, 1], [0, 1], "b--")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.savefig(
    "Images/ROC_Curve_K-Nearest_Neighbours_Classification.png", bbox_inches="tight"
)
```

## Computing the AUC of ROC


```python
print(
    "Area Under the Receiver Operating Characteristic Curve :",
    roc_auc_score(y_test, y_score),
)
```

## Saving the model


```python
dump(knnc, "../../../../Models/k-NearestNeighboursClassifier.joblib")
```
