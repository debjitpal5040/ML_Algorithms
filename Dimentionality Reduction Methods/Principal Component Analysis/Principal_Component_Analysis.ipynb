# Principal Component Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)  # type: ignore
df["target"] = pd.Series(cancer.target)  # type: ignore
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
X_train.shape
```

## Perform PCA for all components first


```python
pca = PCA(n_components=30)
X_pca = pca.fit_transform(X_train)
X_pca
```

```python
a = np.cumsum(pca.explained_variance_ratio_ * 100)
print(a)
```

## Plot the graph explaining cumulative variance


```python
plt.figure(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
plt.plot(a)
plt.grid()
plt.yticks(np.arange(50, 105, 5))
plt.xticks(np.arange(0, 33, 3))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.savefig("Images/Variance_curve.png", bbox_inches="tight")
```

# Taking the components accounting with 95% variability of the dataset


```python
pca_95 = PCA(n_components=0.95)
X_pca_95 = pca_95.fit_transform(X_train)
```

```python
X_pca_95.shape
```

```python
plt.figure(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
plt.plot(X_pca_95)
plt.xlabel("Sample Index")
plt.ylabel("Principal Component")
plt.title("95% Explained Variance")
plt.savefig("Images/Explained_Variance.png", bbox_inches="tight")
```

## New Dataset


```python
df_new = pd.DataFrame(
    X_pca_95,
    columns=[
        "PCA 1",
        "PCA 2",
        "PCA 3",
        "PCA 4",
        "PCA 5",
        "PCA 6",
        "PCA 7",
        "PCA 8",
        "PCA 9",
        "PCA 10",
    ],
)
df_new["target"] = y_train
df_new.head()
```
