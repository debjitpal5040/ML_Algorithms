```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
```

```python
df = pd.read_csv("../Datasets/Anomalies.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
X1 = df["Feature1"].values.reshape(-1, 1)  # type: ignore
X2 = df["Feature2"].values.reshape(-1, 1)  # type: ignore
```

```python
lof = LocalOutlierFactor(n_neighbors=3)
y1 = lof.fit_predict(X1)  # type: ignore
y2 = lof.fit_predict(X2)  # type: ignore
```

```python
print(np.sort(X1.flatten()))
```

```python
print(np.sort(X1[y1 == -1].flatten()))
```

```python
plt.figure(figsize=(8, 5), dpi=150)
plt.scatter(
    X1,
    X1,
    c=y1,
    cmap="coolwarm",
    edgecolor="k",
)
plt.title("Local Outlier Factor on Feature 1")
plt.legend(["Inliers"])
plt.show()
```

```python
print(np.sort(X2.flatten()))
```

```python
print(np.sort(X2[y2 == -1].flatten()))
```

```python
plt.figure(figsize=(8, 5), dpi=150)
plt.scatter(
    X2,
    X2,
    c=y2,
    cmap="coolwarm",
    edgecolor="k",
)
plt.title("Local Outlier Factor on Feature 2")
plt.legend(["Inliers"])
plt.show()
```
