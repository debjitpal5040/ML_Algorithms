# Parametric Density Estimation


```python
from matplotlib import pyplot as plt
from numpy import mean, std
from numpy.random import normal
from scipy.stats import norm
```

```python
sample = normal(loc=50, scale=5, size=200)
sample_mean = mean(sample)
sample_std = std(sample)
print("Actual sample mean:", sample_mean)
print("Actual sample std:", sample_std)
minima = min(sample)
maxima = max(sample)
```

## Visualization of sample


```python
plt.figure(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
plt.hist(sample, density=True)
plt.savefig("Images/Histogram.png", bbox_inches="tight")
```

```python
dist = norm(sample_mean, sample_std)
```

```python
values = [value for value in range(round(minima), round(maxima) + 1)]
probabilities = [dist.pdf(value) for value in values]  # type: ignore
```

```python
plt.figure(figsize=(6, 6), dpi=150)
plt.style.use("dark_background")
plt.hist(sample, density=True)
plt.plot(values, probabilities)
plt.savefig("Images/Histogram_with_PDF.png", bbox_inches="tight")
```
