# Kernel Density Estimation (Parzen Window)

```python
from matplotlib import pyplot as plt
from numpy.random import normal
from numpy import hstack, asarray, exp
from sklearn.neighbors import KernelDensity
```

## Generate a Sample

```python
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = hstack((sample1, sample2))
```

## Plot the Histogram

```python
plt.figure(figsize=(6, 6), dpi=150)
plt.style.use('dark_background')
plt.hist(sample, bins=50)
# plt.xticks("values")
# plt.yticks("probabilities")
plt.title("Histogram")
plt.savefig('Images/Histogram.png', bbox_inches='tight')
```

## Fit Density

```python
model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)
```

## Sample Probabilities for a Range of Outcomes

```python
values = asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)
```

## Plot the Histogram and PDF

```python
plt.figure(figsize=(6, 6), dpi=150)
plt.style.use('dark_background')
plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
# plt.xticks("Values")
# plt.yticks("Probabilities")
plt.title("Histogram with PDF")
plt.savefig("Images/Histogram_with_PDF.png", bbox_inches='tight')
```
